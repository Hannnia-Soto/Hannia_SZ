```python
"""Hannia Soto Zurita, 1948211"""


"""Primer ejercicio"""

def saludar_usuario():
    nombre = input("Introduce tu nombre, por favor: ")

    # Revisamos si el nombre está en la lista de nombres excluidos
    if nombre.lower() in ["ana", "juan", "pedro"]:
        print("Lo siento, no puedo saludarte.")
    else:
        print("Hola, {}".format(nombre))

# Ejecutamos el algoritmo
saludar_usuario()

"""Segundo ejercicio"""

def invertir_lista():
    # El usuario ingresa el núm. de alimentos
    n = int(input("Por favor, ingresa el número de elementos en la lista: "))

    # Pedimos al usuario que ingrese los elementos de la lista
    lista = []
    for i in range(n):
        elemento = input("Ingresa el elemento {}: ".format(i+1))
        lista.append(elemento)

    # Invertimos el orden de la lista
    lista_invertida = lista[::-1]

    # Mostramos la lista invertida al usuario
    print("La lista original es:", lista)
    print("La lista invertida es:", lista_invertida)

# Ejecutamos
invertir_lista()

"""Tercer ejercicio"""

def fibonacci(n):
    # Inicializamos los primeros dos números de la serie de Fibonacci
    fibonacci_series = [0, 1]

    # Generamos los siguientes números de la serie de Fibonacci
    for i in range(2, n):
        fibonacci_series.append(fibonacci_series[-1] + fibonacci_series[-2])

    return fibonacci_series

def generar_fibonacci():
    # Pedimos al usuario que ingrese cuántos números de la serie de Fibonacci desea generar
    n = int(input("Por favor, ingresa cuántos números de la serie de Fibonacci deseas generar: "))

    # Generamos la serie de Fibonacci
    serie = fibonacci(n)

    # Imprimimos la serie de Fibonacci
    print("La serie de Fibonacci con {} números es:".format(n))
    for num in serie:
        print(num)

# Ejecutamos
generar_fibonacci()

"""Cuarto ejercicio"""

import random

def juego_adivinar_numero():
    # Generamos un número aleatorio entre 1 y 100
    numero_objetivo = random.randint(1, 100)

    # Inicializamos el contador de intentos
    intentos = 0

    while True:
        # Pedimos al usuario que ingrese su intento
        intento = int(input("Intenta adivinar el número (entre 1 y 100): "))

        # Incrementamos el contador de intentos
        intentos += 1

        # Comparamos el intento del usuario con el número objetivo
        if intento < numero_objetivo:
            print("El número es mayor. Intenta nuevamente.")
        elif intento > numero_objetivo:
            print("El número es menor. Intenta nuevamente.")
        else:
            print("Felicidades, adivinaste el número {} en {} intentos!".format(numero_objetivo, intentos))
            break

# Ejecutamos
juego_adivinar_numero()

"""Quinto ejercicio"""

def es_primo(numero):
    if numero <= 1:
        return False
    for i in range(2, int(numero**0.5) + 1):
        if numero % i == 0:
            return False
    return True

def numeros_primos_en_rango():
    # Pedimos al usuario que ingrese el rango de números
    inicio = int(input("Ingresa el inicio del rango: "))
    fin = int(input("Ingresa el final del rango: "))

    # Mostramos los números primos dentro del rango dado
    print("Los números primos en el rango de {} a {} son:".format(inicio, fin))
    for numero in range(inicio, fin + 1):
        if es_primo(numero):
            print(numero)

# Ejecutamos
numeros_primos_en_rango()

"""Sexto ejercicio"""

def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def multiplicacion(a, b):
    return a * b

def division(a, b):
    if b != 0:
        return a / b
    else:
        return "Error: No se puede dividir entre cero."

def calculadora_basica():
    # Solicitamos al usuario que ingrese dos números
    num1 = float(input("Ingresa el primer número: "))
    num2 = float(input("Ingresa el segundo número: "))

    # Operaciones
    print("\nOpciones de operaciones disponibles:")
    print("1. Suma")
    print("2. Resta")
    print("3. Multiplicación")
    print("4. División")

    # Solicitamos al usuario que elija una operación
    opcion = input("Por favor, elige una operación (1/2/3/4): ")

    # Realizamos la operación seleccionada por el usuario
    if opcion == '1':
        print("La suma de {} y {} es: {}".format(num1, num2, suma(num1, num2)))
    elif opcion == '2':
        print("La resta de {} y {} es: {}".format(num1, num2, resta(num1, num2)))
    elif opcion == '3':
        print("La multiplicación de {} y {} es: {}".format(num1, num2, multiplicacion(num1, num2)))
    elif opcion == '4':
        print("La división de {} entre {} es: {}".format(num1, num2, division(num1, num2)))
    else:
        print("Esta opción no es válida, ingrese las posibles")

# Ejecutamos
calculadora_basica()







```
